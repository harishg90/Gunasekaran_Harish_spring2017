```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import glob

path = "../Flight data/*_2016.csv"
frame=pd.DataFrame()
df_list=[]

for fname in glob.glob(path):    
    df=pd.read_csv(fname,index_col=None, header=0)
    df_list.append(df)
frame = pd.concat(df_list)
frame.head()
frame=frame.fillna(0)
frame["DEP_HOUR"]=frame["DEP_TIME"].astype(int)
frame["HOUR"]=(frame["DEP_HOUR"]//100)
frame.head()
```

```python
def f(row):
    if row == "1":
        val="MONDAY"
    elif row == "2":
        val="TUESDAY"
    elif row == "3":
        val="WEDNESDAY"
    elif row == "4":
        val="THURSDAY"
    elif row == "5":
        val="FRIDAY"
    elif row == "6":
        val="SATURDAY"
    else:
        val="SUNDAY"
    return val
frame['weekday_name'] = frame.apply(lambda x:f(str(x['DAY_OF_WEEK'])),axis=1)
frame.head()
```

```python
df=DataFrame(frame,columns=["HOUR","weekday_name","ORIGIN_CITY_NAME","FL_NUM"])
offdata=DataFrame(df.groupby(['ORIGIN_CITY_NAME','weekday_name','HOUR'])['FL_NUM'].count())
offdata=offdata.reset_index()
```

```python
city=input("text")
weekday=input("text")
offdata=offdata[offdata.ORIGIN_CITY_NAME==city]
offdata=offdata[offdata.weekday_name==weekday]


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
import time
#saturated=pd.read_csv("output/melted.csv")
sns.factorplot(x="FL_NUM",y="weekday_name",hue="HOUR",data=offdata,saturation=2,size=10,aspect=1,kind="bar",legend=True)

plt.title('No of Flights',fontsize=15,y=1.08)
plt.xlabel('Weekday')
plt.ylabel('HOURLY FLIGHT TOOK OFF FROM ANY CITY')
ax = plt.gca()
ax.get_xaxis().get_major_formatter().set_scientific(False)
plt.draw()
plt.figtext(1.0, 1.10, 'Date Generated:'+str(datetime.datetime.fromtimestamp(time.time()).strftime('%Y-%m-%d %H:%M:%S')), horizontalalignment='right')

fig1 = plt.gcf()
plt.show()
```
