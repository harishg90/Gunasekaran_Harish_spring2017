```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import glob
import requests
import os
import json
import datetime
import time
path_val_fix="Football/Head2Head/"
directory_fix_id= "match_fix.csv"
df_fixtureid=pd.read_csv(directory_fix_id,sep=',')
fixture_list=df_fixtureid.values.tolist()
for i in range(0,len(fixture_list)):
    v=fixture_list[i][1]
    k=fixture_list[i][0]
    head2head_list=requests.get("http://api.football-data.org/v1/fixtures/"+str(v)+"?head2head=50").json()
    teams=k.split(',')
    #print(teams)
    file_name_match =path_val_fix+ teams[0]+'vs'+ teams[1]
    os.makedirs(os.path.dirname(file_name_match), exist_ok=True)
    with open(file_name_match, 'w', encoding='utf-8') as f:
        json.dump(head2head_list, f)
        f.close()
    #print(head2head_list)
    time.sleep(5)
```
