```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import glob
df1=pd.read_csv("September_2016.csv",sep=",") 
df1.head()
pdf=DataFrame(df1, columns=['DEST_CITY_NAME', 'CARRIER_DELAY','WEATHER_DELAY','NAS_DELAY','SECURITY_DELAY','LATE_AIRCRAFT_DELAY'])
pdf['CARRIER_DELAY']=pdf['CARRIER_DELAY'].fillna(0)
pdf['WEATHER_DELAY']=pdf['WEATHER_DELAY'].fillna(0)
pdf['NAS_DELAY']=pdf['NAS_DELAY'].fillna(0)
pdf['SECURITY_DELAY']=pdf['SECURITY_DELAY'].fillna(0)
pdf['LATE_AIRCRAFT_DELAY']=pdf['LATE_AIRCRAFT_DELAY'].fillna(0)
pdf['CARRIER_DELAY'] = pdf['CARRIER_DELAY'].apply(lambda x: 0 if x==0 else 1)
pdf['WEATHER_DELAY'] = pdf['WEATHER_DELAY'].apply(lambda x: 0 if x==0 else 1)
pdf['NAS_DELAY'] = pdf['NAS_DELAY'].apply(lambda x: 0 if x==0 else 1)
pdf['SECURITY_DELAY'] = pdf['SECURITY_DELAY'].apply(lambda x: 0 if x==0 else 1)
pdf['LATE_AIRCRAFT_DELAY'] = pdf['LATE_AIRCRAFT_DELAY'].apply(lambda x: 0 if x==0 else 1)
pdf.to_csv("in.csv")



```

```python
out=DataFrame(pdf.groupby(['DEST_CITY_NAME'])[['CARRIER_DELAY','WEATHER_DELAY','NAS_DELAY','SECURITY_DELAY','LATE_AIRCRAFT_DELAY']].sum())
out=out.reset_index()
out=out.sort_values(by='CARRIER_DELAY',ascending=False)
out.head(n=20).to_csv("output.csv",columns=['DEST_CITY_NAME','CARRIER_DELAY','WEATHER_DELAY','NAS_DELAY','SECURITY_DELAY','LATE_AIRCRAFT_DELAY'])

```

```python
## Graph Analysis for Tip Percentage on a weekday basis

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# import the data directly into a pandas dataframe
nba = pd.read_csv("output.csv", index_col='DEST_CITY_NAME')
# remove index title
nba.index.name = ""
# normalize data columns
nba_norm = (nba - nba.mean()) / (nba.max() - nba.min())
# relabel columns
labels = ['CARRIER_DELAY','WEATHER_DELAY','NAS_DELAY','SECURITY_DELAY','LATE_AIRCRAFT_DELAY']
nba_norm.columns = labels
# set appropriate font and dpi
sns.set(font_scale=1.2)
sns.set_style({"savefig.dpi": 75})
sns.set
# plot it out
ax = sns.heatmap(nba_norm, cmap=plt.cm.Blues, linewidths=1)
# set the x-axis labels on the top
ax.xaxis.tick_top()
ax.set(xlabel='Weekday', ylabel='Hour of Day')
# rotate the x-axis labels
plt.xticks(rotation=90)
# get figure (usually obtained via "fig,ax=plt.subplots()" with matplotlib)
fig = ax.get_figure()
# specify dimensions and save
fig.set_size_inches(10, 15)
fig.savefig("nba.png")
```
