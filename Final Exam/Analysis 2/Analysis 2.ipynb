```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import glob

path = "Flight data/*_2016.csv"
frame=pd.DataFrame()
df_list=[]
airdata = pd.read_csv("Flight data/Airline_LOV.csv", sep=',')[["AIRLINE_ID", "CARRIER_NAME"]]

for fname in glob.glob(path):    
    df=pd.read_csv(fname,index_col=None, header=0)
    df_list.append(df)
frame = pd.concat(df_list)
frame.AIRLINE_ID.count()

```

```python
#To get distance covered by each Airline
df2=DataFrame(frame.groupby('AIRLINE_ID')['DISTANCE'].sum())
df2.reset_index()
```

```python
financedata = pd.read_csv("Flight Data/FinanceData.csv",sep=',')
df4 = DataFrame(financedata[financedata.apply(lambda x: x['QUARTER'] == 3, axis = 1)])
df4 = DataFrame(df4.groupby(["AIRLINE_ID", "CARRIER_NAME"])["NET_INCOME"].sum())

df4.reset_index()
```

```python
df3 = df2.merge(df4, how = 'left', left_index = True, right_index=True)
df3['INCOME_PER_MILE'] = df3['NET_INCOME']/df3['DISTANCE']
df3=df3.reset_index()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

ax = df3['DISTANCE'].plot(kind="bar", alpha=1)
ax2 = ax.twinx()
ax2.plot(ax.get_xticks(),df3['NET_INCOME'],marker='o', c='black', linewidth=4)
ax.set_xlabel("Airlines")
ax.set_ylabel("Distance")
ax.set_title("Distance Covered/Net Income Analysis")
ax.set_xticklabels(df3['CARRIER_NAME'].tolist())
ax.set_ylim(0,df3["DISTANCE"].max())
ax2.set_ylim(0,df3["NET_INCOME"].max())
ax2.grid(False)
fig1 = plt.gcf()
fig=plt.figure(figsize=(10,10))
fig1.savefig("distance_netincome_analysis.png")
plt.show()
```
