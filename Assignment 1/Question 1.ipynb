```python
import nltk
import operator
nltk.corpus.gutenberg.fileids()
words = nltk.corpus.gutenberg.words('austen-persuasion.txt')
len(words)
unique_words=(set(words))
austen_dict={}
for x in unique_words:
    austen_dict[x]=words.count(x)
print(sorted(austen_dict.items(),key=operator.itemgetter(1),reverse=True)[10:])
```
